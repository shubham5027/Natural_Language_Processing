```python
import pandas as pd
import numpy as np
df=pd.read_csv('data.csv')
```

```python
df.head()
```

1) Lowercasing All Text

```python


df['Sentence']=df['Sentence'].str.lower()
```

```python
df.head()
```

2) Removing HTML Tags

```python
import re
def remove_tags(text):
  pattern=re.compile('<.*?>')
  return pattern.sub(r'',text)
```

```python
df['Sentence']=df['Sentence'].apply(remove_tags)
```

```python
df.head()
```

3) Removing URL

```python
import re
def remove_url(text):
  pattern=re.compile(r'https?://\S+|www\.S+')
  return pattern.sub(r'',text)
```

```python
df['Sentence']=df['Sentence'].apply(remove_url)
```

```python
df.head()
```

4) Removing Punctuation

```python
import string

exclude=string.punctuation

def remove_punct(text):
  return text.translate(str.maketrans('','',exclude))

df['Sentence']=df['Sentence'].apply(remove_punct)

```

```python
df.head()
```

Spell Checker

```python
from textblob import TextBlob

def spell_check(text):

    txtblb = TextBlob(text)
    return str(txtblb.correct())

df['column_text'] = df['column_text'].apply(spell_check)
```

Removing Emmmogies

```python
import re

def remove_emojis(text):

    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F700-\U0001F77F"  # alchemical symbols
        u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        u"\U0001FA00-\U0001FA6F"  # Chess Symbols
        u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE,
    )
    return emoji_pattern.sub(r"", text)

df['column_name'] = df['column_name'].apply(remove_emojis)
```

Removing stop words

```python
import nltk
from nltk.corpus import stopwords
import pandas as pd

nltk.download('stopwords')

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = nltk.word_tokenize(text)
    filtered_words = [word.lower() for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)


df['text_column'] = df['text_column'].apply(remove_stopwords)
```

Tokenization using Spacy

```python
import spacy
import pandas as pd


nlp = spacy.load("en_core_web_sm")

def spacy_tokenize(text):
    doc = nlp(text)
    tokens = [token.text for token in doc]
    return tokens

df['text_column'] = df['text_column'].apply(spacy_tokenize)
```

Tokenization using NLTK

```python
import nltk
import pandas as pd

nltk.download('punkt')

def tokenize_text(text):

    tokens = nltk.word_tokenize(text)
    return tokens

df['text_column'] = df['text_column'].apply(tokenize_text)
```

Lemmitization using Spacy

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def spacy_lemmatize(text):
    doc = nlp(text)
    lemmatized_tokens = [token.lemma_ for token in doc]
    return lemmatized_tokens

df['text_column'] = df['text_column'].apply(spacy_lemmatize)
```

Stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')

def nltk_stem(text):
    stemmer = PorterStemmer()
    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)

df['text_columns'] = df['text_column'].apply(nltk_stem)
```
