```python
import re
import string
import nltk
import spacy
from textblob import TextBlob
import pandas as pd
from spacy.lang.en.stop_words import STOP_WORDS
from nltk.stem import PorterStemmer
nltk.download('punkt')
# from spellchecker import SpellChecker
nltk.download('stopwords')
nlp = spacy.load("en_core_web_sm")
# spell_checker = SpellChecker()

def preprocess_text(text):
    # Lowercasing
    text = text.lower()

    # Remove HTML tags and URLs in one step
    text = re.sub(r'<.*?>|https?://\S+|www\S+', '', text)

    # Remove punctuation
    text = ''.join(char for char in text if char not in string.punctuation)


    # Spell check using pyspellchecker
    # tokens = nltk.word_tokenize(text)
    # text = [spell_checker.correction(token) for token in tokens]


    # Spell check using TextBlob  #Time Consuming
    # txtblb = TextBlob(text)
    # text = str(txtblb.correct())

    # Remove emojis
    # emoji_pattern = re.compile(
    #     "["
    #     u"\U0001F600-\U0001F64F"
    #     u"\U0001F300-\U0001F5FF"
    #     u"\U0001F680-\U0001F6FF"
    #     u"\U0001F700-\U0001F77F"
    #     u"\U0001F780-\U0001F7FF"
    #     u"\U0001F800-\U0001F8FF"
    #     u"\U0001F900-\U0001F9FF"
    #     u"\U0001FA00-\U0001FA6F"
    #     u"\U0001FA70-\U0001FAFF"
    #     u"\U00002702-\U000027B0"
    #     u"\U000024C2-\U0001F251"
    #     "]+",
    #     flags=re.UNICODE,
    # )
    # text = emoji_pattern.sub(r"", text)

    # Porter stemming (3 sec)
    stemmer = PorterStemmer()
    tokens = nltk.word_tokenize(text)
    stemmed_tokens = [stemmer.stem(token) for token in tokens if token.lower() not in STOP_WORDS]


    #Lemmitization  (50 sec)
    # lemmatized_tokens = [token.lemma_ for token in nlp(text) if token.lemma_.lower() not in STOP_WORDS]

    # Combine tokens into a single string  (for stemming)
    processed_text_combined = ' '.join(stemmed_tokens)

    # Combine tokens into a single string  (for lemmitzing)
    # processed_text_combined = ' '.join(lemmatized_tokens)


    return processed_text_combined

df['processed_data'] = list(nlp.pipe(df['Message'], batch_size=2000, n_process=4,disable=['parser']))
df['processed_data'] = df['Message'].apply(preprocess_text)
```
